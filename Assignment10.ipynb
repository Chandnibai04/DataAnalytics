# List all cities where the area is greater than 5000, along with their population.

```python
# Query to get cities with population > 5000
query_1 = """
SELECT name, population 
FROM cities 
WHERE population > 5000;
"""

# Execute and display the result
sql_query(query_1)
```

# List the names of capital cities where the death rate is greater than 2% and the
# birth rate is less than 20%

```python
query_2 = """
SELECT cities.name 
FROM cities
JOIN facts ON cities.facts_id = facts.id
WHERE cities.capital = 1 
AND facts.death_rate > 2 
AND facts.birth_rate < 20;
"""

sql_query(query_2)
```

# List the cities with a population growth rate between 1% and 5%, sorted in desc
# by population growth.

```python
query_3 = """
SELECT cities.name, facts.population_growth
FROM cities
JOIN facts ON cities.facts_id = facts.id
WHERE facts.population_growth BETWEEN 1 AND 5
ORDER BY facts.population_growth DESC;
"""

sql_query(query_3)
```

# Find countries that start with "A" and do not have any cities listed in the cities
# table. The query should return the country name and city name. Using left join

```python
query_4 = """
SELECT facts.name AS country_name, cities.name AS city_name
FROM facts
LEFT JOIN cities ON facts.id = cities.facts_id
WHERE facts.name LIKE 'A%' AND cities.name IS NULL;
"""

sql_query(query_4)
```

# List all capital cities and their corresponding death rate using a LEFT JOIN.

```python
query_5 = """
SELECT cities.name AS capital_city, facts.death_rate
FROM cities
LEFT JOIN facts ON cities.facts_id = facts.id
WHERE cities.capital = 1;
"""

sql_query(query_5)
```
